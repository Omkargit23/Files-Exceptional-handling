**Theory Questions :**

**1. What is the difference between interpreted and compiled languages?**
  Ans-
Interpreted language:
Code is executed line by line.
Example: Python

Compiled language:
Code is converted into machine code before running.
Example: C, C++

Difference:
Interpreted → slower, easier to debug
Compiled → faster, runs directly on machine


**2. What is exception handling in Python?**
  Ans - Exception handling is used to handle errors in a program without stopping it.

It uses try and except blocks to catch and manage errors


**3. What is the purpose of the finally block in exception handling?**
  Ans - The finally block is used to run code no matter what happens.

It executes whether an error occurs or not (used for cleanup like closing files).


**4. What is logging in Python?**
  Ans - Logging in Python is used to record messages about a program’s execution (like errors or info).

It helps in debugging and tracking what the program is doing.


**5. What is the significance of the __del__ method in Python?**
  Ans- The __del__ method is a destructor in Python.

It is called when an object is about to be destroyed, usually used for cleanup (like releasing resources).


**6. What is the difference between import and from ... import in Python?**
  Ans-
import:
Imports the whole module.

from ... import:
Imports specific parts from a module.

Difference:
import → use module name
from ... import → use directly without module name


**7. How can you handle multiple exceptions in Python?**
  Ans- You can handle multiple exceptions using multiple except blocks or in one line.

Example:
try:
    x = int("abc")
except ValueError:
    print("Value error")
except TypeError:
    print("Type error")


**8. What is the purpose of the with statement when handling files in Python?**
  Ans- The with statement is used to open files safely.

It automatically closes the file after use, even if an error occurs.


**9. What is the difference between multithreading and multiprocessing?**
  Ans-
  Difference:
Multithreading:
Runs multiple threads in the same process (shares memory).
Used for lightweight tasks.

Multiprocessing:
Runs multiple processes (separate memory).
Used for heavy tasks.

```python
# 2. Write a Python program to read the contents of a file and print each line?

file = open("file.txt", "r")

for line in file:
    print(line.strip())

file.close()
```

```python
# 3. How would you handle a case where the file doesn't exist while trying to open it for reading?

try:
    file = open("file.txt", "r")
    print(file.read())
    file.close()
except FileNotFoundError:
    print("File not found")
```

```python
# 4. Write a Python script that reads from one file and writes its content to another file?

with open("file1.txt", "r") as source:
    content = source.read()

with open("file2.txt", "w") as target:
    target.write(content)
```

```python
# 8. Write a program to handle a file opening error using exception handling?

try:
    file = open("file.txt", "r")
    print(file.read())
    file.close()
except FileNotFoundError:
    print("File not found")
except Exception:
    print("Error opening file")

```

```python
# 9. How can you read a file line by line and store its content in a list in Python?

lines = []

try:
    with open("file.txt", "r") as file:
        for line in file:
            lines.append(line.strip())

    print(lines)

except FileNotFoundError:
    print("File not found")
```

```python
# 10. How can you append data to an existing file in Python?

with open("file.txt", "a") as file:
    file.write("New data\n")
```

```python
# 13. How would you check if a file exists before attempting to read it in Python?

import os

if os.path.exists("file.txt"):
    with open("file.txt", "r") as file:
        print(file.read())
else:
    print("File does not exist")
```

```python
# 14. Write a program that uses the logging module to log both informational and error messages?

import logging

logging.basicConfig(level=logging.DEBUG)

logging.info("This is an info message")

try:
    x = 10 / 0
except ZeroDivisionError:
    logging.error("Error occurred: division by zero")
```

```python
# 15. Write a Python program that prints the content of a file and handles the case when the file is empty?

try:
    with open("file.txt", "r") as file:
        content = file.read()

        if content == "":
            print("File is empty")
        else:
            print(content)

except FileNotFoundError:
    print("File not found")

```
